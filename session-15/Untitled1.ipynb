```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

from sklearn.model_selection import train_test_split

from PIL import Image

from sklearn.cluster import KMeans
```

```python
img = Image.open("../datasets/strange.jpg")
```

```python
small = img.resize(size=(400, 300))
```

```python
small
```

```python
img_arr = np.array(small)
```

```python
plt.imshow(img_arr)
```

```python
img_arr.shape
```

```python
X = img_arr.reshape(-1, 3)
```

```python
X.shape
```

```python
model = KMeans(n_clusters=10)
```

```python
model.fit(X)
```

```python
pallets = model.cluster_centers_.astype(np.uint8)
```

```python
pallets
```

```python
plt.imshow(pallets.reshape(1, 10, 3))
```

```python
model.labels_[:100]
```

```python
output = pallets[model.labels_]
```

```python
output.shape
```

```python
out_img = output.reshape(*img_arr.shape)
```

```python
plt.imshow(out_img)
```
